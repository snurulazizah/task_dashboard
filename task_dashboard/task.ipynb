```python
import requests

url = "https://app.qontak.com/api/v3.1/tasks"
params = {
    "filter": "alltask",   # string
    "page": 1,             # numeric
    "per_page": 25         # numeric
    }
headers = {
    "Authorization": "68d419f1a1953c9196c9257856cf31fb15f679119e0c6b6e96a8729a57ceece5",  # ganti dengan token dari Qontak
    "Accept": "application/json"
}

response = requests.get(url, headers=headers, params=params)

if response.status_code == 200:
    tasks = response.json()
    print(tasks)
else:
    print("Error:", response.status_code, response.text)
```
